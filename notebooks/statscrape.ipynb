```python
import pandas as pd
import requests as r
```

```python
response = r.get('https://statsapi.mlb.com/api/v1/game/718572/playByPlay')
```

```python
german = []

for play in response.json()['allPlays']:
    if play['matchup']['pitcher']['id'] == 593334:
        german.append(play)
```

```python
german
```

```python
german[0]
```

```python
df = []

for play in german:
    for pitch in play['playEvents']:
        if 'call' in pitch['details'].keys():
            row = {
                'inning': play['about']['inning'],
                'pitch': pitch['pitchNumber'],
                'call': pitch['details']['description'],
                'pitch_type': pitch['details']['type']['code'],
                'speed': pitch['pitchData']['startSpeed'],
                'rpm': pitch['pitchData']['breaks']['spinRate']
            }

            df.append(row)
```

```python
pd.DataFrame(df).to_csv("~/desktop/german.csv")
```
